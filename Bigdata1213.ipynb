```python
# 1 Mengidentifikasi dan Menangani Data yang Hilang
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24.0, 30.0, None, 22.0],
    'Salary': [48000.0, 57000.0, 57000.0, None]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())


print('After cleaning:\n', df)
```

```python
#2 Standarisasi Data Kategorikal
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standarisasi nilai kategori
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Muat dataset
df = pd.read_csv('Marine_Fish_Data.csv')

# Identifikasi kolom numerik
kolom_numerik = df.select_dtypes(include=['float64', 'int64']).columns
print("Kolom numerik:", kolom_numerik)

# Normalisasi kolom numerik
scaler = MinMaxScaler()

if len(kolom_numerik) > 0:
    df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])
    print("Data setelah normalisasi:")
    print(df.head())  # Menampilkan 5 baris pertama dari dataset yang dinormalisasi
else:
    print("Tidak ada kolom numerik untuk dinormalisasi.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Muat dataset
df = pd.read_csv('Marine_Fish_Data.csv')

# Menghitung jumlah ikan berdasarkan Region
grouped_data = df['Region'].value_counts().head(10)  # Ambil 10 region teratas
labels = grouped_data.index
values = grouped_data.values

# Menghitung jumlah ikan berdasarkan Breeding Season
grouped_data2 = df['Breeding_Season'].value_counts().head(10)  # Ambil 10 musim pemijahan teratas
values2 = grouped_data2.values

# Menyesuaikan panjang labels jika berbeda
min_length = min(len(labels), len(grouped_data2.index))
labels = labels[:min_length]
values = values[:min_length]
values2 = values2[:min_length]

x = np.arange(len(labels))  # posisi x untuk setiap grup
width = 0.35  # lebar batang

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, values, width, label='Region', color='brown')
bars2 = ax.bar(x + width/2, values2, width, label='Breeding Season', color='black')

# Menambahkan label, judul, dan legenda
ax.set_ylabel('Jumlah')
ax.set_title('Perbandingan Jumlah Ikan dalam Region dan Breeding Season')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Muat dataset
df = pd.read_csv('Marine_Fish_Data.csv')

# Batasi dataset hanya pada 100 entri teratas
limited_df = df.head(100)  # Mengambil 100 entri pertama

# Membuat plot sebar
plt.figure(figsize=(5, 3))
sns.scatterplot(data=limited_df, x='Average_Size(cm)', y='Region', hue='Region', palette='viridis', style='Region', s=100)

# Menambahkan judul dan label
plt.title('Plot Sebar: Hubungan antara Ukuran Rata-rata Ikan dan Wilayah (100 Ikan Teratas)')
plt.xlabel('Ukuran Rata-rata (cm)')
plt.ylabel('Wilayah')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
